```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
data=pd.read_csv("D:\CA_251090050530\mtcars - Copy.csv")
print(data.info())
print("Stats for miles per gallon:")
print("Mean:",data["mpg"].mean())
print("Median  :",data["mpg"].median())
print("Min   :",data["mpg"].min())
print("Max  :",data["mpg"].max())
print("Stats for horsepower:")
print("Mean:",data["hp"].mean())
print("Median  :",data["hp"].median())
print("Min   :",data["hp"].min())
print("Max  :",data["hp"].max())
print("Stats for weight:")
print("Mean:",data["wt"].mean())
print("Median  :",data["wt"].median())
print("Min   :",data["wt"].min())
print("Max  :",data["wt"].max())
plt.plot(data["cyl"],data["mpg"])
plt.xlabel("Cylinders")
plt.ylabel("Efficiency")
plt.show()
plt.scatter(data["hp"],data["mpg"])
plt.xlabel("HORSEPOWER")
plt.ylabel("MPG")
plt.title("RELATION BETWEEN HP AND MPG")
plt.show()
a=data[data['am']==1]['mpg']
m=data[data['am']==0]['mpg']
plt.figure()
plt.boxplot([a,m])
plt.xticks([1,2],["Automatic","Manual"])
plt.xlabel("Transmission type")
plt.ylabel("Mpg")
plt.show()
sn.scatterplot(data=data,x=data["hp"],y=data["mpg"],hue='model')
print("Therefore the car with high efficiency as well as high hp is Ferrari Dino as it lies in the upper right quadrant of the scatter plot.We can also say that as hp increases mpg decreases.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
df=pd.read_csv("D:\CA_251090050530\cereal.csv")
missing_values=df.isnull().sum()
print("Missing values before treatment:",missing_values)
print(df.info())
print("Avg calories:",df["calories"].mean())
print("Avg sugar:",df["sugars"].mean())
print("Avg fibre content:",df["fiber"].mean())
sn.barplot(x=df["name"],y=df["sugars"])
plt.subplots_adjust(bottom=0.2)
plt.show()
sn.barplot(x=df["calories"],y=df["rating"])
plt.show()
sn.scatterplot(data=df,x=df["calories"],y=df["fat"],hue='name')
plt.show()
print("Therefore we can make out that according to the calories and fat scatter plot All-Bran with extra fiber is the healthiest cereal.It is concluded that the least calories cereal gets highest rating.")
```
